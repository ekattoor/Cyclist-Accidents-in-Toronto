## Preparing the data for Machine learning

```python
cyclists = cyclists.drop(columns=['YEAR','DATE'])
```

```python
cyclists.columns.values
```

```python
#setting dummy values for categorical values
ROAD_CLASS_mv = pd.get_dummies(cyclists['ROAD_CLASS'], prefix="ROADCLASS", prefix_sep ='_')
ACCLOC_mv = pd.get_dummies(cyclists['ACCLOC'], prefix="ACCLOC", prefix_sep ='_')
TRAFFCTL_mv = pd.get_dummies(cyclists['TRAFFCTL'], prefix="TRAFFCTL", prefix_sep ='_')
VISIBILITY_mv = pd.get_dummies(cyclists['VISIBILITY'], prefix="VISIBILITY", prefix_sep ='_')
LIGHT_mv = pd.get_dummies(cyclists['LIGHT'], prefix="LIGHT", prefix_sep ='_')
RDSFCOND_mv = pd.get_dummies(cyclists['RDSFCOND'], prefix="RDSFCOND", prefix_sep ='_')
RDSFCOND_mv = pd.get_dummies(cyclists['RDSFCOND'], prefix="RDSFCOND", prefix_sep ='_')
ACCLASS_mv = pd.get_dummies(cyclists['ACCLASS'], prefix="ACCLASS", prefix_sep ='_')
IMPACTYPE_mv = pd.get_dummies(cyclists['IMPACTYPE'], prefix="IMPACTYPE", prefix_sep ='_')
INVTYPE_mv = pd.get_dummies(cyclists['INVTYPE'], prefix="INVTYPE", prefix_sep ='_')
District_mv=pd.get_dummies(cyclists['District'], prefix="District", prefix_sep ='_')
VEHTYPE_mv = pd.get_dummies(cyclists['VEHTYPE'], prefix="VEHTYPE", prefix_sep ='_')
DRIVCOND_mv = pd.get_dummies(cyclists['DRIVCOND'], prefix="DRIVCOND", prefix_sep ='_')
```

```python
cyclists['SPEEDING'] =cyclists['SPEEDING'].astype('category').cat.codes
cyclists['ALCOHOL'] =cyclists['ALCOHOL'].astype('category').cat.codes
cyclists['HOUR'] =cyclists['HOUR'].astype('category').cat.codes
```

```python
#Renaming injury as Injury =1 and minor or no injury as 0
cyclists['INJURY'].replace(to_replace='None', value=0, inplace=True)
cyclists['INJURY'].replace(to_replace='Major', value=1, inplace=True)
cyclists['INJURY'].replace(to_replace='Fatal', value=1, inplace=True)
cyclists['INJURY'].replace(to_replace='Minimal', value=0, inplace=True)
cyclists['INJURY'].replace(to_replace='Minor', value=0, inplace=True)
```

```python
#concatenating 

cyclists_mv = pd.concat([cyclists, ROAD_CLASS_mv,District_mv,ACCLOC_mv,TRAFFCTL_mv,VISIBILITY_mv,LIGHT_mv,
       RDSFCOND_mv, ACCLASS_mv, IMPACTYPE_mv, INVTYPE_mv,
       VEHTYPE_mv,DRIVCOND_mv], axis=1)
```

```python
#dropping the original columns
cyclists_mv.drop(['Index_','District','LOCCOORD','INVAGE','AG_DRIV','ROAD_CLASS','ACCLOC', 'TRAFFCTL', 'VISIBILITY', 'LIGHT',
       'RDSFCOND', 'ACCLASS', 'IMPACTYPE', 'INVTYPE',
       'VEHTYPE', 'DRIVCOND'],axis=1,inplace=True)
```

```python
#displaying after setting dummy variables
cyclists_mv.head(10)
```

```python
#Checking for null values
cyclists_mv.isnull().sum()
```

## Split the data into X and y and then into testing and training sets

```python
#Slicing the data as target value and feature values
X = cyclists_mv.drop(columns=['INJURY'])
Y = cyclists_mv["INJURY"]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

```python
#Splitting the data as training data and testing data
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=.23, random_state=7) 
```

# Fit a Logistic Regression Classifier

```python
from sklearn.linear_model import LogisticRegression

logistic_regression_model = LogisticRegression(random_state=7)
```

```python
# Fit and save the logistic regression model using the training data
lr_model = logistic_regression_model.fit(X_train, y_train)
```

## Create the predicted values for the testing and the training data.

```python
#Generate training predictions
training_predictions = lr_model.predict(X_train)

#Generate testing predictions
testing_predictions = logistic_regression_model.predict(X_test)
```

## Print a confusion matrix for the training data

```python
# Import the model for sklearn's confusion matrix
from sklearn.metrics import confusion_matrix

# Create and save the confustion matrix for the training data
training_matrix = confusion_matrix(y_train, training_predictions)

# Print the confusion matrix for the training data
print(training_matrix)
```

## Print a confusion matrix for the texting data

```python
# Create and save the confustion matrix for the testing data
test_matrix = confusion_matrix(y_test, testing_predictions)

# Print the confusion matrix for the testing data
print(test_matrix)
```

## Print the training classification report.

```python
# Create and save the training classifiction report
training_report = classification_report(y_train, training_predictions)

# Print the training classification report
print(training_report)
```

## Print the testing classification report

```python
# Create and save the testing classifiction report
testing_report = classification_report(y_test, testing_predictions)

# Print the testing classification report
print(testing_report)
```
